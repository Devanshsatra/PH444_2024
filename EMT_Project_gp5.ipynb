# **Localized charge imaging with scanning Kelvin probe microscopy**

#**1. INTRODUCTION**

Electrical scanning force microscopy related techniques, such as electrostatic force microscopy and scanning Kelvin probe microscopy (SKPM), offer the possibility of obtaining the topography and the electrical information simultaneously at the nanoscale with the advantage of not being limited to conducting samples . Due to the complexity of the tip-sample system, to make the technique quantitative, it is necessary to deal with issues related to the optimum implementation and the ultimate resolution of the technique, the best description of the probe geometry and the correct data interpretation . Nowadays, SKPM is a very precise tool that has been successfully applied to an enormous variety of 1D and 2D organic and inorganic nanostructures. The local work function, the doping profile or the dielectric constant are consistently obtained in metallic, semiconducting and dielectric samples, respectively .

The expansion of the SKPM characterization to new materials includes systems in which localized charges are present, such as organic and inorganic adsorbates on dielectric substrates  disordered hopping systems , point and edge defects or electrostatic nanopatterns. While for a conducting tip-sample system the electrostatic interaction has been studied thoroughly, in presence of localized charges the quantitative interpretation of the Kelvin voltage (VSKPM) is still under debate . Qualitatively, a localized/trap charge appears as a potential domain in the SKPM image but the amount of charge and the charge spatial distribution cannot be directly inferred and additional theoretical modelling is needed. The Kelvin voltage due to localized charges depends on the tip–sample distance, tip geometric parameters (tip radius, tip cone angle, etc) and on the relative permittivity $(ε_r = ε/ε_0)$ of the charge surrounding medium . Moreover, there are no analytical expressions and even simple geometries are addressed numerically. Several progresses have been achieved in this direction; point charges close to a metallic surface and large dipolar islands were modeled with the spherical tip approximation. More realistic probe geometries have been employed in numerical simulations of insulator-supported molecules and large charge domains nanopatterned on polymethyl methacrylate (PMMA).

In the present work, an easily-handled intuitive approach based on the method of image charges is proposed to model the SKPM images in thick dielectric films with small localized charges. We analyze the Kelvin voltage, discussing the quantitative data interpretation depending on the SKPM measurement mode. First, the Kelvin voltage due to a point charge located at the surface of a semi-infinite dielectric is simulated. Then, the problem is generalized, to study small charge domains. To model the tip, a spherical geometry is used, neglecting the cone and cantilever contributions. With this simplified tip, we study the tip–sample distance Kelvin voltage dependence, as well as the lateral resolution during image acquisition. To validate the method, experimental measurements performed on charge domains of different sizes injected in PMMA discontinuous films are compared with our theoretical predictions. From the Kelvin voltage images the lateral size and the total stored charge of the charge domains is extracted and the predictions are confirmed by spectroscopy curves. In addition, the effects of the real probe geometry are discussed, and the most adequate measurement parameters are determined.

# **2. Theoretical modelling**



In the absence of localized charges, the energy associated to the electrostatic interaction between a biased metallic tip$ (V_{bias})$ and a grounded substrate is described by

where VCP = Δϕ/e is the classical surface contact potential associated with the difference of the tip and sample work functions (Δϕ) and e is the electron charge, $C(z, ε_r, ζ_i)$ is the capacitance of the tip-sample system which depends on the tip–sample distance (z), the εr of the sample and on the geometric parameters of the tip (ζi) such as the tip apex radius (R), the cone angle (θ), tip height (H), etc. The corresponding electrostatic force is

$$
F_{elec} = \frac{1}{2}\frac{dC(z,ϵ_r,ζ_i)}{dz}(V_{bias} - V_{CP})^2
$$

For small frequency shifts and small oscillation amplitudes, the shift in resonance frequency due to the electrostatic interaction $(Δf_{elec}) $and the force gradient are related by

$$
\frac{Δf_{elec}}{f_0} = -\frac{1}{2k}\frac{dF}{dz} = -\frac{1}{4k}\frac{d^2C(z,ϵ_r,ζ_i)}{dz^2}(V_{bias} - V_{CP})^2
$$

where $f_0$ is the free resonance frequency. When a localized charge q is present, additional terms should be included in equation (1) in order to take into account the charge-induced polarization in the tip and in the sample. The electrostatic energy is

$$
W_{elec}(V_{bias}) = u_0(z,ϵ_r,ζ_i,q) + u_1(z,ϵ_r,ζ_i,q) ×(V_{bias} - V_{CP}) + u_2(z,ϵ_r,ζ_i) (V_{bias} - V_{CP})^2 .
$$

The polarization generates a potential that is independent on$ (V_{bias}−V_{CP}) $and interacts with the charge itself described by the term $u_0$. The term $u_1$ is due to the interaction of the charge with the potential generated by the capacitor and is proportional to $ (V_{bias}−V_{CP}) $. Finally the term $u_2$ =$ −C(z, ε_r, ζ_i)/2 $ is the energy of the capacitor, which is independent of the localized charges. The corresponding electrostatic force and frequency shifts are then given by:

$$
F_{elec} = \frac{du_0}{dz} + \frac{du_1}{dz} (V_{bias}−V_{CP})  + \frac{du_2}{dz} (V_{bias}−V_{CP}) ^2
$$
$$
\frac{Δf_{elec}}{f_0} = - \frac{1}{2k}\frac{dF}{dz}\\= -\frac{1}{2k}(\frac{d^2u_0}{dz^2} + \frac{d^2u_1}{dz^2}(V_{bias}−V_{CP}) +\frac{d^2u_2}{dz^2} (V_{bias}−V_{CP} )^2)
$$

Note, that both $F_{elec}$ and$ Δf_{elec}/f_0 $ still have a parabolic dependence on the bias.

The calculation of the terms $u_0, u_1, u_2$ is not straightforward and strongly depends on the tip geometry and on the dielectric relative permittivity. For metallic substrates $(ε_r = ∞)$, the capacitance related term, $u_2$, has been widely modelled for different tips geometries. Approximated analytical expressions have been obtained for a spherical metallic tip and for more realistic probe geometries which include the cone and the cantilever contributions . For dielectric samples, the capacitance of a sphere near a thick dielectric layer has been widely studied numerically. For thin dielectric films and certain tip–sample distances, approximated analytical expressions have been provided generalizing the expression obtained for the metallic case . The localized charge related terms  $u_0$ and  $u_1$ should be calculated numerically, and only few specific system geometries have been addressed.

# **2.1. Image charge method: sphere-point change model**

In this work, we use the well established method of images to calculate the $F_{elec}$ and $ Δf_{elec}/f_0$ in the presence of small localized charges . First, we will discuss the simple case of a point charge located at the dielectric surface $(z = 0)$ of a semi-infinite dielectric film $(ε_r)$. Note that when a charge q is placed on a dielectric, it is automatically screened being the total effective charge $q_0 = (1-s)q$ with $s = (ε_r − 1)/(ε_r + 1)$. As schematically shown in figure 1, in our model, the tip consists of a metallic spherical tip of radius R at a fixed voltage $(V_{bias})$. In the method of images the Vbias is substituted by the potential of a point charge $(Q1 = 4πε_0RV_{bias})$ located at the centre of the sphere, which induces an infinite set of successive image charges inside the dielectric
 and in the metallic sphere $(Q_n)$ represented with black dots in figure 1. All these charges are proportional to Vbias. In the same way, the localized point charge at the surface $(q_0 = (1-s)q)$ generates an infinite set of successive image charges inside the sphere $(q_n)$ and in the dielectric $(q_n' = −sq_n)$. The Felec acting on the sphere is the sum of all forces between charges inside the sphere and charges outside the sphere. Note that in this approximation the image charges at the metal back electrode are neglected and therefore this method is only valid for 'thick' dielectric films compared with the tip radius dimensions, than is h ≫ R. For 'thick' films our results are independent on the dielectric thickness that can be considered as semi-infinite.

### **2.2. SKPM contrast and lateral resolution**


Both SKPM contrast and lateral resolution are ultimately dependent on the electromagnetic interactions between the electron beam and the sample.
The electron beam's properties, like energy and spot size, are controlled by electromagnetic fields within the SEM column.
SKPM is a variation of Scanning Electron Microscopy (SEM) that utilizes a vibrating probe tip to scan a sample surface. The work function, which is the energy required to remove an electron from a material's surface, varies across the surface of most samples. In SKPM, the vibrating probe tip comes into close proximity to the sample surface, but not close enough for physical contact. The work function difference between the tip and the sample creates a contact potential difference (CPD) between them. This CPD modulates the vibration amplitude of the probe tip, which is then detected. SKPM contrast relies on the work function, which is an inherent electromagnetic property of the material, which modulates the probe vibration for image formation. Variations in the vibration amplitude correspond to SKPM contrast, providing information on the sample's work function distribution.

In section 2 we have seen  \
$F_{elec}$ = $\frac{du_0}{dz} + \frac{du_1}{dz} (V_{bias} - V_{CP}) + \frac{du_2}{dz} (V_{bias} - V_{CP})^2$

The minima of this parabolic equation gives us the following two results

$V^{AM}_{SKPM}$ = $(-\frac{1}{2} \frac{\frac{du_1}{dz}}{\frac{du_2}{dz}} + V_{CP})$
----- if AM-SKPM is used. And

$V ^{FM}_{SKPM}$ = $(-\frac{1}{2} \frac{\frac{d^2u_1}{dz^2}}{\frac{d^2u_2}{dz^2}} + V_{CP})$
----- if FM-SKPM is used.

To simplify, we'll assume $V_{CP}=0$ (contact potential) since its contribution merely serves as an offset. This would benefit us as our focus will be directed towards the relationship between $V_{SKPM}$ and the localized charges. Then our results change to \

$V^{AM}_{SKPM}$ = $(-\frac{1}{2} \frac{\frac{du_1}{dz}}{\frac{du_2}{dz}})$
And
$V ^{FM}_{SKPM}$ = $(-\frac{1}{2} \frac{\frac{d^2u_1}{dz^2}}{\frac{d^2u_2}{dz^2}})$.

Now we plot $V ^{AM}_{SKPM}$ and $V ^{FM}_{SKPM}$, respectively, against the tip-sample distance atop the point charge q for various dielectric media. This is depicted in Figures 3(a) and (b).

Fig. 3. (a) $V ^{AM}_{SKPM} (z, ε_r, q=1, R=1, x=0)$ and (b) $V ^{FM}_{SKPM} (z, ε_r, q=1, R=1, x=0)$ as a function of distance for a point charge placed on a thick dielectric film of $ε_r = 2.5, 3, 4.1, 5, 7, 10$ and $15$.


As anticipated, increasing the dielectric constant $(ε_r)$ results in a decrease in the Kelvin voltage due to heightened charge screening. Generally, both values exhibit disparity, with $V ^{AM}_{SKPM}$ and $V ^{FM}_{SKPM}$ displaying dependence on $z$ at short tip-sample distances $(z < R)$, with this dependency being more pronounced in the $V ^{FM}_{SKPM}$ signal. As the tip-sample distance becomes large $(z > R)$, they converge towards a common offset value, which can be analytically determined in the limit as $z → ∞$, which gives us following result.

$V ^{charge}_{SKPM}$ = $ \frac{2(1-s)q}{4\pi ε_0 s R} $

If R and $ε_r$ are known, the total charge q can be determined from the offset of the $V_{SKPM}$ at long distances (as z → ∞) observed in both $V ^{AM}_{SKPM}$ and $V ^{FM}_{SKPM}$ measurements. However, it's important to note that in the current model, the contributions from the cone and cantilever are not considered. These factors may significantly influence the results, particularly at large tip-sample distances.



To analyze the formation of contrast in SKPM images and the apparent lateral size of the point charge in both $V ^{AM}_{SKPM}$ and $V ^{FM}_{SKPM}$ modes, Kelvin voltage lateral profiles have been computed at various tip-sample distances. This can be seen plotted in below figure.

When comparing the lateral profiles of $V ^{AM}_{SKPM}$ and $V ^{FM}_{SKPM}$, figures 4(a) and (b) respectively, it is evident that the $V ^{FM}_{SKPM}$ signal exhibits slightly superior lateral resolution compared to $V ^{AM}_{SKPM}$, which corroborates previously found expression of $\Delta \frac{f_{elec}}{f_0}$. Moreover, even at typical tip-sample distances operated in air, the presence of a localized charge can influence the Kelvin voltage value, extending its effect to large lateral distances from the charge.




Hence to summarise, It is concluded that in the presence of localized point charges, the lateral resolution is not solely determined by the tip radius and tip-sample distance, as in any SFM-related technique, but also by the values of $R, ε_r,$ and $z$. This implies that the resolution of the technique is contingent upon the specific system. For instance, having a $+1e$ charge at $z = 1R$ with $R = 20 nm$ in a material with $εr = 3$ results in $V_{SKPM} = 185 mV$, whereas in a material with $ε_r = 10$, it yields only $V_{SKPM} = 43 mV$. From an experimental standpoint, this suggests that appropriate selection of sharp tips and small tip-sample distances is crucial to ensure a sufficiently large detectable signal. Furthermore, utilizing the $V^{FM}_{SKPM}$ mode enhances lateral resolution.








### 2.3. Finite size charge domain



1. **Introduction to Charge Domain and Localization Length**:
- In many systems, the charge is not concentrated at a single point but rather spread over a region, characterized by a length known as the localization length. This section explores how the Kelvin voltage depends on the size of this charge domain, modeled as a disc with homogeneous charge distribution.

2. **Generalization of Image Charges Model**:
- The paper mentions that extending the image charges model used for point charges to a spread density of charge is straightforward. The online supporting information likely contains the detailed mathematical derivation for this extension.

3. **Dependence of Kelvin Voltage on Charge Domain Size**:
- Figures 5(a) and (b) in the paper depict the Kelvin voltage $V_{SKPM}^{AM} and V_{SKPM}^{FM}$ as a function of tip-sample distance in the center of the charge domain for different charge disc radii.


fig5_P.jpg


   - At large distances, the behavior converges to the point charge limit. Total charge can be obtained from an equation provided in the paper.
   - At small and medium distances, two types of curves are observed based on the relative domain size:
     - For domain sizes smaller than the tip radius $(r < R)$, the behavior is point-charge like.
     - For domain sizes larger than the tip radius $(r \geq R)$, the behavior is different. $V_{SKPM}$ decreases until it converges to a constant value.
   - The paper provides an analytical expression for $V_{SKPM}$ in the limit as $z \rightarrow 0$, which depends only on domain size, permittivity, and charge density.
   $$
      V_{SKPM} = \frac{(1-s)q}{2πε_\circ r} = \frac{(1-s)σr}{2ε_\circ}
   $$
   where $\sigma = \frac{q}{\pi r^2}$

- Figures 6(a) and (b) illustrate the apparent lateral domain size and $V_{SKPM}$ contrast for a fixed permittivity and tip-sample distance, calculated for different charge domain radii.

5. **Practical Considerations**:
- To analyze Kelvin voltage accurately and extract quantitative information about stored charge and domain size, parameters such as tip radius, relative permittivity, and tip-sample distance need to be characterized.
- Lateral resolution and $V_{SKPM}$ contrast increase at short tip-sample distances, with higher resolution in FM-SKPM mode.
- Sub-tip-radius charge domains can be resolved even at specific distances.
- The model assumes the tip is a sphere, neglecting other contributions like cone and cantilever, which may be relevant at larger tip-sample distances.

# ExperimentaI Setup

In this section, we validate the theoretical model through experimental means. We'll examine the optimal parameters for data collection, address any limitations encountered during the experiments, and propose potential solutions. Specifically, we derive the total charge and the size of charge domains from experimental Scanning Kelvin Probe Microscopy (SKPM) images. These images were captured on small, localized charge domains introduced onto densely packed Poly(methyl methacrylate) (PMMA) through contact electrification. The sample comprises a non-continuous PMMA film on highly ordered pyrolytic graphite (HOPG), as depicted in Figure 7(a). Additional information regarding sample preparation is available in the online supporting materials. This sample setup offers several advantages: (1) HOPG regions serve as a reference for metal during measurements, (2) the precise thickness of the PMMA film (typically 200–300 nm) is determined from topography images, and (3) the surface charge nanopatterning on PMMA using Scanning Force Microscopy (SFM) is a well-established technique documented in literature. The corresponding SKPM image (Figure 7(b)) prior to any charge injection demonstrates a uniform Kelvin voltage distribution in HOPG regions, while PMMA displays small charge domains, characteristic of surfaces of amorphous, poorly conducting polymers. The associated capacitance image (Figure 7(c)) indicates that regardless of the PMMA thickness, ranging from 200–350 nm, all PMMA regions exhibit the same contrast, i.e., identical capacitance. This suggests that, under the current experimental imaging conditions, the PMMA film behaves akin to a thick film.



- **Experimental methodology**

The experiments were conducted under standard room temperature and ambient conditions utilizing a Nanotec Scanning Force Microscopy (SFM) system equipped with a Phase-Locked Loop (PLL) dynamic measurement board. Platinum-coated tips (OMCLAC240TM-R3) with a specified stiffness of k = 3Nm−1 were employed. To minimize thermal drift, the microscope setup was allowed to stabilize for approximately 24 hours prior to conducting the relevant measurements. For ensuring accurate quantitative measurements and achieving high lateral resolution, topography images were acquired in frequency modulation non-contact dynamic mode (FMDSFM), maintaining an oscillation amplitude of 1 nm and tip-to-sample distances ranging from 7–10 nm (approximately 0.5 times the tip radius). Voltage-Sensing Kelvin Probe Microscopy (VSKPM) FM images were obtained using FM-SKPM mode with a bias voltage of Vac = 500 mV at an electrical modulation frequency of ωelec = 7 kHz. Further details regarding the SKPM setup can be found elsewhere. A 3D voltage spectroscopy mode, as described in the supplementary information and relevant literature, was employed to acquire experimental u2″(z) and u1″(z) curves, which represent the capacitance and charge-related interaction contributions, respectively, as functions of the tip-to-sample distance. It is emphasized that prior to any local charge characterization, precise knowledge of the tip geometry, dielectric constant (εr) of the material, and exact tip-to-sample distance during image acquisition is crucial. Therefore, the experiments were conducted in a three-step sequence: firstly, the tip geometry (including apex radii (R) and cone angle (θ)) was characterized using the HOPG metal reference. Secondly, the dielectric constant of PMMA was determined. Finally, small charge domains were patterned on the PMMA surface and characterized using FM-SKPM imaging and 3D spectroscopy curves. Experimental and theoretical data were then compared to deduce the lateral size of the domains and the total stored charge.

neglected. On the contrary, in air conditions, due to the
hydrophilic character of the sample surface or due to
technical limitations, typical tip-sample working distances are
much larger and the contributions of the microscopic
part of the tip may not be ruled out. To overcome this
problem, in order to simplify the data modelling and to
obtain quantitative information of the localized charges,
instead of the $ V_{FM}^{SKPM} $ the $u_1^{''}(z, q, 𝛆_r, R)$
signal is more adequate. Experimentally,$ V_{FM}^{SKPM} (x, y, z = z_0; q, 𝛆_r, R)$
and the $u_2^{''}(x, y, z = z_0; 𝛆_r, R)$ images are recorded
simultaneously by using a two channel lock-in [28].
Then the $u_1^{''}(x, y, z = z_0; 𝛆_r, R)$ image is reconstructed
($u_1^{''} = -2 V_{FM}^{SKPM}*u_2^{''} )$. Finally, it should be noted that the
present $ V_{SKPM} $ image analysis, only works if FM-KPM mode
is used. If a similar analysis is performed on  $ V_{SKPM}^{AM} $
images of small charge domains, it is not possible to accurately fit
neither the lateral profiles nor the spectroscopy curves with
the simple spherical tip model. This indicates that the contributions of the microscopic parts of the tip are relevant not
only in the capacitance related term but also in the $ u_{1}^{'} $ term.

# Conclusion
In this work we have proposed a methodology to analyse
SKPM images when small localized charges are present on
thick dielectric surfaces in order to obtain quantitative information about the total stored charge and the charge lateral
size. A rigorous protocol to characterize the tip geometrical
parameters and the $ε_r$ of the dielectric material is presented,
since the Kelvin voltage drastically depends on them independently of the SKPM mode used (AM-SKPM or FM-SKPM). Operating the SKPM at short enough tip–sample
distances in the FM-SKPM mode, the $V_{SKPM}^{FM}$ images can be
accurately analysed neglecting the tip cone contribution with
a simple spherical tip image charges-based model. Otherwise,
the contribution of the microscopic cone should be taken into
account, which appreciably increases the numerical effort and
the data analysis renders more difficult. It has been found that
the cone effect mainly contributes the $V_{SKPM}^{FM}$ though the
capacitance related $ u_{2}^{''}(z, 𝛆_r, R, θ)$ term, while the charge
dependent $ u_{1}^{''}(z, q, 𝛆_r, R, θ)$ term is less sensitive to it.
Hence, to avoid the cone effect in ambient conditions, where
the operating tip–sample distances are typically large, we
propose to use the $ u_{1}^{''}(x, y, z, q, 𝛆_r, R)$ signal to obtain more
quantitative results.
Finally, we highlight that the theoretical model proposed
can be easily implemented to model complex surface charges
distributions. Indeed, in order to determine the charge distribution of a general system from Kelvin voltage images, the
inverse problem should be solved. This is a challenging task,
but the simplicity of the image charges method together with
the evidence that the $ u_{1}^{''}(z, q, 𝛆_r, R)$term can be correctly
modelled with the spherical tip approximation even at relative
large tip–sample distances, suggests that this is not out of the
reach and will be addressed in a future work
