HISTORICAL SURVEY AND MAIN PURPOSES (210260056)

Intermolecular forces have been studied for approximately
three centuries. Because molecules of a real gas condense
into liquids and freeze into solids, it is natural to expect that
there exist attractive intermolecular forces—a conclusion
that had already been reached by Newton at the end of the
17th century. The phenomenon of capillarity—the ability of
a liquid to climb the walls of a tube in opposition to external
forces like gravity—was studied for the first time by Clairaut
who, in 1743, suggested that this phenomenon could be
explained if the forces between the molecules of the liquid
and those of a tube of glass were different from the intermolecular forces between the molecules of the liquid themselves.This same phenomenon was considered later by
Laplace, in 1805, and by Gauss, in 1830. Many others,
including as Maxwell and Boltzmann, were also involved
with the study of intermolecular forces. A more complete list
of contributors to this subject up to the 20th century can be
found in Israelachvili’s book.<br>

Following a different approach, van der Waals suggested
in his dissertation in 1873 an equation of state for real gases,
given for one mole of gas by **(P +
a/V^2
)(V − b) = RT,**
where P, V, and T are, respectively, the pressure, volume,
and absolute temperature of the gas, R is the universal gas
constant, and a and b are two adjustable parameters.
Parameter b was introduced to take into account the finite
volumes of the molecules, while the term a/V^2 is related to
the existence of an attractive intermolecular force. These
attractive forces are called van der Waals forces. In fact, we
must distinguish three types of van der Waals forces: the orientation force, the induction force, and the dispersion force,
to be described below.<br>

Orientation forces occur between two molecules with
permanent electric dipoles. Taking a thermal average of the
electrostatic interaction between two randomly oriented
dipoles of moments p<sub>1</sub> and p<sub>2</sub>, Keesom computed the van
der Waals interaction energy between two polar molecules
in a thermal bath at (absolute) temperature T and obtained
$$ U_{\text{or}}(r) = -\frac{2p_1^2 2p_2^2}{3k_BT (4\pi\varepsilon_0)^2 r^6} ; \text{ for } k_BT \gg \frac{p_1 p_2}{4\pi\varepsilon_0 r^3}
  (1)$$

where p<sub>1</sub>= |p<sub>1</sub>|, p<sub>2</sub> = |p<sub>2</sub>|,
r is
the distance between the two molecules, and k<sub>B</sub> is the
Boltzmann constant. The minus sign means that the interaction is attractive. Though there are as many configurations
that give rise to attractive forces as configurations that give
rise to repulsive forces,   the Boltzmann factor
(e^-E/k<sub>B</sub>T)favors the lower energies that correspond to “attractive configurations.” Note that as T increases indefinitely all configurations become equally available, leading to a vanishing
force.<br>
Induction forces occur between a non-polar but polarizable
molecule and another one that possesses a permanent electric
dipole (or a higher multipole, as an electric quadrupole).<br>


Evidence that non-polar molecules indeed existed led Debye
and others to consider this kind of force. The permanent dipole
of one molecule induces a dipole in the non-polar but polarizable one, leading to a behavior similar to the previous dipoledipole interaction. If **p<sub>1</sub>** is the magnitude of the dipole moment
of the polar molecule, then the magnitude of the electric field
at the position of the non-polar but polarizable one will be
**E<sub>1</sub>(2)≈p<sub>1</sub>/r^3** and the magnitude of the induced dipole
acquired by this molecule will be **p<sub>2</sub>=α<sub>2</sub>E<sub>1</sub>(2),**
 α<sub>2</sub> being its
static polarizability. Apart from a numerical factor, the interaction energy is **U<sub>ind</sub>≈ -p<sub>2</sub>E<sub>1</sub>(2) ≈ -α<sub>2</sub>p<sub>1</sub>^2/r^6**. The induction
force does not disappear at high temperatures because the orientations of dipoles 1 and 2 are not independent. Indeed, the
induced dipole is parallel (for isotropic molecules) to the field
generated by dipole 1 at the position of molecule 2, which
explains the attractive character of the induction van der Waals
interaction.
<br>
The above two types of van der Waals forces do not
explain the attraction between two atoms or non-polar molecules, like those of noble gases. The explanation for this kind
of force had to wait for the advent of quantum mechanics.
Due to quantum fluctuations, the charge and current distributions in an atom fluctuate and, consequently, instantaneous
dipoles exist and give rise to an electromagnetic interaction.
These fluctuations are ultimately related to the Heisenberg
uncertainty principle. In 1930, Eisenschitz and London and
London considered the interaction between two hydrogen
atoms in detail. After using a simple perturbative method,
they related the interaction potential between the atoms directly to the atomic polarizability of the hydrogen.
Because the dynamical polarizability **α(ω)** is related to the
permittivity **ε(ω)**, these forces were called by London dispersion van der Waals forces.The dispersion interaction
energy between two atoms can be written as
$$ U_{\text{disp}}(r) = -\frac{3}{4} \frac{\hbar \bar{\omega}_0 \alpha_0^2}{(4\pi\varepsilon_0)^2 r^6}
(2)$$
where ω<sub>0</sub>
 is the dominant transition frequency for the interaction and  α<sub>0</sub>
 is the static polarizability of the atoms. A quantum mechanical derivation of London's result (2) can be found in standard textbook
<br>
Experiments with colloids performed at the Phillips laboratories in the first half of the 1940s by Verwey and Overbeek
disagreed with London’s prediction. These experiments
showed that the dispersion interaction between two atoms falls
for large distances more rapidly than 1/r^6. Further, Overbeek
conjectured that such a change in the force law was due to retardation effects of the electromagnetic interaction. These
effects become important as the time required for light to
propagate from one atom to the other is comparable to the
characteristic times of the atoms, namely, 1/ω<sub>mn</sub>, where ω<sub>mn</sub>
are the allowed atomic transition frequencies. Assuming there
is a dominant transition frequency, say **ω<sub>0</sub>**, retardation effects
become relevant for r/c>=1/ω<sub>0</sub>. We distinguish two regimes
for dispersion interactions: the non-retarded or short distance
regime, and the (asymptotically) retarded or large distance regime. The latter is valid for r>>λ<sub>0</sub> while the former is valid
for a<sub>0</sub><< r << λ<sub>0</sub>, with λ<sub>0</sub> being the dominant transition wavelength and a<sub>0</sub> being the Bohr radius ( a<sub>0</sub><< r avoids
the overlapping of the electronic clouds of the two atoms).
The influence of retardation effects on the London-van der
Waals forces was first reported by Casimir and Polder in the
late 1940s. After a fourth-order perturbative calculation in
QED, they showed that, for r >> λ<sub>0</sub> , the dispersion interaction
energy between two atoms is
$$ U_{\text{ret}}(r) = -\frac{23\hbar c}{4\pi}  \frac{ \alpha_1 \alpha_2}{(4\pi \varepsilon_0)^2 r^7}
(3)$$
where α<sub>1</sub> and α<sub>1</sub> are the static polarizabilities of atoms 1 and 2,
respectively. Note the change in the power law 1/r^6 to 1/^r7.<br>
In contrast to the Coulomb interaction, which obeys the
superposition principle, van der Waals interactions are not pairwise additive, as first noticed by Axilrod and Teller. This fact
must be taken into account in the computation of the van der
Waals force between an atom and a macroscopic body (or
between two macroscopic bodies). A pairwise integration with
London or Casimir and Polder forces would be justified only
for rarefied bodies. Non-additivity effects on the energy of a
system may be positive or negative and are usually small
(less than 20%), but they can be very important, as in the way the
atoms of rare gases are arranged in solids. For more details on
the non-additivity of dispersion forces see the textbooks of
Margenau and Kestner, Langbein, and Milonni. Calculations of the dispersion van
der Waals interaction between two atoms at any separation can
be found in the pedagogical paper by Holstein and in some
textbooks such as those by Craig and Thirunamachandran
and by Salam.<br>
Though we shall not discuss any experiment on dispersion
forces between atoms and macroscopic surfaces, we shall mention a few of them. In 1993, a remarkable experiment
was performed by Sukenik et al.,
 in which for the first time
the change in the power law between retarded and nonretarded regimes was observed directly with atoms. In 1996,
Landragin et al. measured the van der Waals force in an
atomic mirror based on evanescent waves. In 2001, quantum
reflection was used to measure dispersion forces by
Shimizu. A short but valuable description of these experiments can be found in the nice paper by Dalibard.<br>
Dispersion forces appear not only in different areas of
physics, such as atomic and molecular physics, condensed
matter physics, and quantum field theory (QFT) but also in
engineering, chemistry, and biology. In QFT, the dispersion force is closely connected to the Casimir effect; the
history of this subject is well described in introductory
papers and books. There are even more bizarre
situations where dispersion forces play an important role,
like in the adhesion of geckos  or as an important element
in the generation of electric potentials in thunderstorms. A
vast list of references can be found in Milton’s Resource
Letter and in a paper by Buhmann and Welsch. The history of intermolecular forces along the last centuries can be
found in Rowlinson’s excellent book, which contains a
huge bibliography.<br>
In this paper, we shall be concerned only with nonretarded dispersion forces, which do not demand the quantization of the electromagnetic field. In fact, non-retarded
dispersion forces can be computed with ordinary quantum
mechanics. In particular, we shall focus our attention to the
van der Waals interaction between an atom and a perfectly
conducting surface. Our purposes are the following: (i) to
popularize a simple but powerful method proposed in 2007
by Eberlein and Zietal  that is extremely well suited for this
kind of calculation (and used afterwards by them and other
authors,); (ii) to show that the image method of electrostatics can be extremely useful in applying the Eberlein and
Zietal method; and (iii) to present an original result—the calculation of the van der Waals force between an atom and a
conducting “boss hat” (a conducting hemisphere attached to
an infinite conducting plane).<br>
In 1932, Lennard-Jones  was the first to use the image
method to obtain the Hamiltonian interaction operator to be
used in the quantum mechanical calculation of the van der
Waals interaction energy between a polarizable atom and a
conducting surface. Among other things, Lennard-Jones was
interested in the instantaneous adsorption of gases on the
outer surface of a solid. With this motivation in mind, he
computed the van der Waals interaction energy between an
atom and a perfectly conducting (infinite) plane and found
an attractive interaction energy proportional to 1/z^3, where z
is the distance from the atom to the plane. However, in the
(asymptotically) retarded regime, Casimir and Polder later
showed that the (attractive) interaction energy between an
atom and an infinite conducting plane is proportional to 1/z^4,
rather than 1/z^3. It is worth mentioning that in LennardJones’ result a factor of 1/2 is missing. This numerical error
follows from a misuse of the image method. Recall that, for
the case at hand, this method states that the force acting on a
dipole near a perfectly conducting (infinite) plane is that
exerted by the image dipole and, consequently, the electrostatic
energy between the dipole and the conducting plane is given
by 1/2 times the interacting energy between the dipole and its
image. A detailed but elementary explanation of the presence
of this factor of 1/2 can be found in the literature. A similar calculation missing the same numerical factor occurs in the
quantum mechanics textbook of Cohen-Tannoudji, Diu, and
Laloe. Almost a decade after the Lennard-Jones paper,
Bardeen and Margeneau and Pollard presented more realistic models for computing the interaction energy between an
atom and a metallic surface that took into account the structure
of the metal and enabled them, among other things, to analyze
deviations from the results obtained by the image method.
(For the interaction of a neutral but polarizable atom with
a dielectric surface see the paper by Mavroyannis, published
in 1963).<br>
The non-retarded interaction between an atom and two
parallel and perfectly conducting plates can also be obtained
with the aid of the image method; in this case, there will be
an infinite number of images. A general discussion, including any distance regime, can be found in Barton’s paper.
The image method can still be used in the calculation of the
van der Waals force between an atom and a non-dispersive,
homogeneous, and isotropic medium. For instance, for the
case of an atom near a semi-infinite slab filled with such a
medium, the van der Waals force on the atom is the same as
if the atom were near a perfectly conducting plate except for
an extra factor of (ε-1)/(ε+1), where ε is the electric permittivity of the dielectric medium. We should mention that
in Ref.  the authors consider the non-retarded interaction
between an atom in an arbitrary level and a dielectric surface
characterized by a frequency dependent, complex, dielectric
constant. Finally, the van der Waals force between an atom
and a perfectly conducting sphere (grounded or isolated) was
computed recently by Taddei et al. Analogous results for
metallic and dielectric nanospheres were obtained by
Schmidt et al.; however, for simplicity, we shall be concerned here only with perfectly conducting bodies <br>
Having completed this historical summary, we next
review the Eberlein and Zietal method, showing how we can
combine it with the image method. In Sec. III, we will illustrate the method by solving a couple of introductory problems. In Sec. IV, we treat the nontrivial case of an atom near
a conducting boss hat which, to the best of our knowledge,
has not appeared in the literature before.

In this section, we briefly review Eberlein and Zietal's method. This method has the advantage of relating the quantum problem to a classical one in electrostatics in a systematic and enlightening way, mainly when the corresponding classical problem admits a simple solution by the image method.

In order to compute the van der Waals interaction between two neutral but polarizable atoms, say two hydrogen atoms, we first need to find out the interaction Hamiltonian operator between them to be used in the perturbative quantum-mechanical calculation of their interaction. In this case, the unperturbed Hamiltonian operator of the system formed by the two atoms is simply the sum of the Hamiltonian operators of both atoms separately, which already include the Coulomb interactions existing in both atoms. Because we are in the short distance regime (non-retarded regime), this is achieved by computing the Coulomb interaction between all charges of one atom and all charges of the other. Once the distance between them, though small to make retardation effects negligible, is much greater than the Bohr radius, a Taylor expansion can be performed, leading to a dipole-dipole interaction as the dominant term.

However, for the case of a neutral but polarizable atom interacting with a macroscopic body, say a perfectly conducting body, we have to take into account a huge number of pairwise Coulomb interactions. In this case, it is convenient to introduce the electrostatic potential $$\phi$$ and write the Coulomb interaction Hamiltonian between the atom and the conducting body in terms of the electrostatic potential

# IV. ATOM CLOSE TO A CONDUCTING BOSS HAT SURFACE

This surface comprises a conducting hemisphere with radius \( R \) accompanied by an infinite conducting plane. This configuration, known as a hemispherical boss, presents intriguing challenges in electrostatics. To analyze this setup effectively, we utilize the concept of image charges, which play a crucial role in modeling the electric field distribution and potential near such surfaces. The interplay between the atom and the conducting surface in this geometry leads to intricate electrostatic phenomena that merit thorough investigation.

# V. CONCLUSIONS

The method introduced by Eberlein and Zietal (reference 38) offers a practical approach to compute van der Waals interactions between an atom and a conducting surface of any shape. This method has been applied to various scenarios, including simple cases and more complex scenarios like an atom near a conducting boss-hat surface.

One advantage of this method is its ability to quickly transform the problem into a classical electrostatics problem, allowing the use of established methods for solving Laplace or Poisson equations. Particularly, when an image solution is feasible, the van der Waals force can be readily obtained.

The use of the image method is not obligatory; other approaches, such as the Green function method, have been employed for certain scenarios. Nonetheless, the Eberlein and Zietal method combined with the generalized image method, introduced by Sommerfeld, has proven effective even for intricate geometries.

Due to its effectiveness and simplicity, the Eberlein and Zietal method serves as an excellent tool for students learning about non-retarded dispersion forces. It can be extended to handle various systems, providing insight into complex interactions between atoms and macroscopic bodies.

While the focus of this paper is on perfectly conducting bodies, the method can be generalized for scenarios involving non-dispersive dielectrics with appropriate image charge determinations. However, it's important to note that there are other methods available for computing dispersion forces, such as the Lifshitz method and its generalizations.

Dispersion forces remain a subject of active research, especially in the quest for repulsive forces, which are significant in diverse contexts. Notably, past studies have explored repulsive Casimir forces and interactions between electrically and magnetically polarizable atoms. Recent research has also delved into dispersion interactions in various mediums, leading to proposals for experiments involving dispersive forces, including those with Bose-Einstein condensates near periodic gratings.
